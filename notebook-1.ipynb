```python
class Student:
    def __init__(self):
        self.__name = None
        self.__roll_no = None
        self.__cgpa = None

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_roll_no(self, roll_no):
        self.__roll_no = roll_no

    def set_cgpa(self, cgpa):
        if 0 <= cgpa <= 10:
            self.__cgpa = cgpa
        else:
            raise ValueError("CGPA must be between 0 and 10.")

# Create an instance of the Student class
student = Student()

# Set the attributes using the setter methods
student.set_name("John Doe")
student.set_roll_no(101)
student.set_cgpa(9.5)

# Access the private attributes to display them
print(f"Name: {student._Student__name}")
print(f"Roll No: {student._Student__roll_no}")
print(f"CGPA: {student._Student__cgpa}")
```

```python
class Vehicle:
    def __init__(self, vehicle_type, max_speed):
        self.vehicle_type = vehicle_type
        self.max_speed = max_speed
    
    def display_info(self):
        print(f"Vehicle Type: {self.vehicle_type}")
        print(f"Maximum Speed: {self.max_speed} km/h")

class Car(Vehicle):
    def __init__(self, max_speed):
        super().__init__("Car", max_speed)

class Bike(Vehicle):
    def __init__(self, max_speed):
        super().__init__("Bike", max_speed)

class Truck(Vehicle):
    def __init__(self, max_speed):
        super().__init__("Truck", max_speed)

# Create instances of each vehicle type
car = Car(200)
bike = Bike(100)
truck = Truck(120)

# Display information for each vehicle
print("Car Info:")
car.display_info()
print("\nBike Info:")
bike.display_info()
print("\nTruck Info:")
truck.display_info()
```

```python
from math import pi

class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height

# Create instances of different shapes
circle = Circle(5)
rectangle = Rectangle(4, 6)
triangle = Triangle(3, 8)

# Calculate and display areas
print(f"Circle area: {circle.area():.2f}")
print(f"Rectangle area: {rectangle.area()}")
print(f"Triangle area: {triangle.area()}")
```

```python
class BankAccount:
    def __init__(self, account_number):
        self._balance = 0
        self._account_number = account_number
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self._balance:.2f}")
        else:
            raise ValueError("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Withdrawn ${amount:.2f}. New balance: ${self._balance:.2f}")
        else:
            raise ValueError("Invalid withdrawal amount or insufficient funds")
            
    def get_balance(self):
        return self._balance

class SavingsAccount(BankAccount):
    def __init__(self, account_number, interest_rate=0.02):
        super().__init__(account_number)
        self.__interest_rate = interest_rate
    
    def withdraw(self, amount):
        min_balance = 100  # Minimum balance requirement
        if self._balance - amount >= min_balance:
            super().withdraw(amount)
        else:
            raise ValueError(f"Must maintain minimum balance of ${min_balance}")
    
    def add_interest(self):
        interest = self._balance * self.__interest_rate
        self.deposit(interest)
        print(f"Added interest: ${interest:.2f}")

class CurrentAccount(BankAccount):
    def __init__(self, account_number, overdraft_limit=1000):
        super().__init__(account_number)
        self.__overdraft_limit = overdraft_limit
    
    def withdraw(self, amount):
        if amount > 0 and (self._balance - amount) >= -self.__overdraft_limit:
            self._balance -= amount
            print(f"Withdrawn ${amount:.2f}. New balance: ${self._balance:.2f}")
        else:
            raise ValueError(f"Amount exceeds overdraft limit of ${self.__overdraft_limit}")

# Create and test accounts
savings = SavingsAccount("SAV001")
current = CurrentAccount("CUR001")

# Test deposits
savings.deposit(1000)
current.deposit(500)

# Test withdrawals
try:
    savings.withdraw(800)  # Should succeed
    current.withdraw(800)  # Should succeed (uses overdraft)
    savings.withdraw(200)  # Should fail (minimum balance)
except ValueError as e:
    print(f"Error: {e}")

# Test interest
savings.add_interest()
```

```python
class Book:
    def __init__(self, title, author, isbn):
        self.__title = title
        self.__author = author
        self.__isbn = isbn
        self.__is_available = True
        
    def get_title(self):
        return self.__title
        
    def get_author(self):
        return self.__author
        
    def get_isbn(self):
        return self.__isbn
        
    def is_available(self):
        return self.__is_available
        
    def set_availability(self, status):
        self.__is_available = status

class Member:
    def __init__(self, member_id, name):
        self._member_id = member_id
        self._name = name
        self._borrowed_books = []
        
    def borrow_limit(self):
        pass
        
    def can_borrow(self):
        return len(self._borrowed_books) < self.borrow_limit()
        
    def borrow_book(self, book):
        if self.can_borrow() and book.is_available():
            self._borrowed_books.append(book)
            book.set_availability(False)
            print(f"{self._name} has borrowed {book.get_title()}")
        else:
            print("Cannot borrow book - limit reached or book unavailable")

class StudentMember(Member):
    def borrow_limit(self):
        return 2  # Students can borrow up to 2 books

class FacultyMember(Member):
    def borrow_limit(self):
        return 5  # Faculty can borrow up to 5 books

# Create some books and members
book1 = Book("Python Programming", "John Smith", "ISBN001")
book2 = Book("Data Structures", "Jane Doe", "ISBN002")
book3 = Book("Algorithms", "Alan Turing", "ISBN003")

student = StudentMember("S001", "Alice Cooper")
faculty = FacultyMember("F001", "Dr. Robert Brown")

# Test the system
print("Testing Library Management System:")
student.borrow_book(book1)
student.borrow_book(book2)
student.borrow_book(book3)  # Should fail - student limit reached

faculty.borrow_book(book3)
```
